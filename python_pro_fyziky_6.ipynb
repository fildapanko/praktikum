```python
def kmity(x,y):
    # y = [souradnice, rychlost]
    return [y[1],-y[0]]
y0 = [1,0]
e_span = np.linspace(0,10,101)
rkmity = solve_ivp(kmity, (0,10), y0, t_eval=e_span)
print(rkmity)
```

```python
plt.plot(e_span, rkmity.y[0,:])
plt.plot(e_span, rkmity.y[1,:])
```

```python
def pad(x,y):
    return [y[1],-9.81]
y0 = [10,0]
e_span = np.linspace(0,2,101)
rkmity = solve_ivp(pad, (0,2), y0, t_eval=e_span)
plt.plot(e_span, rkmity.y[0,:])
plt.plot(e_span, rkmity.y[1,:])
plt.grid()
```

```python
def projektil(x,y):
    return [y[1], -9.81, y[3],0]

y0 = [0,10,0,0]

x_span = np.linspace(0,2,101)

delo = solve_ivp(projektil, (0,10), y0, t_eval=x_span)

plt.plot(x_span, delo.y[0,:])
plt.grid()
```
